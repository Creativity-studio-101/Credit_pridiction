```python
import numpy as np
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection  import train_test_split
```

```python
aj=pd.read_csv(r"D:\30 Cource\CreditRisk.csv")
aj.head(5)
```

```python
aj.drop(['Loan_ID'],axis=1)
```

```python
aj.isnull().sum()
```

```python
#Null Values
aj.Gender=aj.Gender.fillna("Male")
aj.Married=aj.Married.fillna("Yes")
aj.Dependents=aj.Dependents.fillna(1.0)
aj.Self_Employed=aj.Self_Employed.fillna("No")
aj.LoanAmount=aj.LoanAmount.fillna(aj.LoanAmount.mean())
aj.Loan_Amount_Term=aj.Loan_Amount_Term.fillna(aj.Loan_Amount_Term.mean())
aj.Credit_History=aj.Credit_History.fillna(0)

#Text to numrical
aj.Gender=aj.Gender.replace({'Male':0,'Female':1})
aj.Married=aj.Married.replace({'No':0,'Yes':1})
aj.Education=aj.Education.replace({"Graduate":0,"Not Graduate":1})
aj.Self_Employed=aj.Self_Employed.replace({"No":0,"Yes":1})
aj.Property_Area=aj.Property_Area.replace({"Urban":0,"Semiurban":1,"Rural":2})
aj.Loan_Status=aj.Loan_Status.replace({"Y":1,"N":0})



aj=aj.drop(['Loan_ID'],axis=1)
```

```python
aj.isnull().sum()
```

```python
aj
```

```python
train_aj,test_aj=train_test_split(aj,train_size=0.8,test_size=0.2)
train_aj.head()
```

```python
x_train=train_aj.iloc[:,0:-1]
y_train=train_aj.iloc[:,-1]

x_test=test_aj.iloc[:,0:-1]
y_test=test_aj.iloc[:,-1]
```

#1. Scikit-learn Models

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, accuracy_score, classification_report, confusion_matrix
```

```python
###                           Linear Regression                           ###

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
predictions = model.predict(x_test)
predictions
```

```python
# Calculate evaluation metrics
r2 = r2_score(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)

print(f'R² Score: {r2}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

```python
###                             Decision Tree                             ###

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(x_train, y_train)
predictions = model.predict(x_test)
predictions
```

```python
# Calculate evaluation metrics
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)
conf_matrix = confusion_matrix(y_test, predictions)

# Print results
print(f'Accuracy Score: {accuracy}')
print('Classification Report:\n', report)
print('Confusion Matrix:\n', conf_matrix)
```

```python
###                         Random Forest                               ###

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(x_train, y_train)
Ran_predictions = model.predict(x_train)
predictions
```

```python
# Predictions on the test set
Ran_predictions = model.predict(x_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, Ran_predictions)
report = classification_report(y_test, Ran_predictions)
conf_matrix = confusion_matrix(y_test, Ran_predictions)

# Print results
print(f'Accuracy Score: {accuracy}')
print('Classification Report:\n', report)
print('Confusion Matrix:\n', conf_matrix)
```

```python
# Splitting dataset into train-test
train_aj, test_aj = train_test_split(aj, train_size=0.8, test_size=0.2, random_state=42)
```

```python
# Define feature columns
feature_columns = [col for col in aj.columns if col not in ['Loan_Status', 'LoanAmount']]
```

```python
# Features & target for classification
X_train = train_aj[feature_columns]
y_train = train_aj['Loan_Status']

X_test = test_aj[feature_columns]
y_test = test_aj['Loan_Status']
```

```python
# Train Random Forest Classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
# Classification Accuracy
y_pred = clf.predict(X_test)
print("Loan Approval Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Train Regression Models for Loan Amount Prediction
approved_loans = train_aj[train_aj['Loan_Status'] == 1]
rejected_loans = train_aj[train_aj['Loan_Status'] == 0]

reg_approved = RandomForestRegressor(random_state=42)
reg_rejected = RandomForestRegressor(random_state=42)

reg_approved.fit(approved_loans[feature_columns], approved_loans['LoanAmount'])
reg_rejected.fit(rejected_loans[feature_columns], rejected_loans['LoanAmount'])
```

```python
# Define interactive widgets
gender = widgets.Dropdown(options=[("Male", 0), ("Female", 1)], description="Gender:")
married = widgets.Dropdown(options=[("No", 0), ("Yes", 1)], description="Married:")
dependents = widgets.IntSlider(min=0, max=5, value=1, description="Dependents:")
education = widgets.Dropdown(options=[("Graduate", 0), ("Not Graduate", 1)], description="Education:")
self_employed = widgets.Dropdown(options=[("No", 0), ("Yes", 1)], description="Self-Employed:")
applicant_income = widgets.IntText(value=5000, description="Applicant Income:")
coapplicant_income = widgets.IntText(value=2000, description="Coapplicant Income:")
loan_term = widgets.IntSlider(min=12, max=480, value=360, step=12, description="Loan Term:")
credit_history = widgets.Dropdown(options=[("No", 0), ("Yes", 1)], description="Credit History:")
property_area = widgets.Dropdown(options=[("Urban", 0), ("Semiurban", 1), ("Rural", 2)], description="Property Area:")
requested_loan = widgets.IntText(value=100000, description="Requested Loan Amount:")
predict_button = widgets.Button(description="Check Eligibility")
# Function to predict loan status
def predict_loan_status(_):
    user_data = {
        "Gender": gender.value,
        "Married": married.value,
        "Dependents": dependents.value,
        "Education": education.value,
        "Self_Employed": self_employed.value,
        "ApplicantIncome": applicant_income.value,
        "CoapplicantIncome": coapplicant_income.value,
        "Loan_Amount_Term": loan_term.value,
        "Credit_History": credit_history.value,
        "Property_Area": property_area.value,
    }
    
    user_df = pd.DataFrame([user_data])
    user_df = user_df.reindex(columns=feature_columns, fill_value=0)
    
    loan_status = clf.predict(user_df)[0]
    
    if loan_status == 1:
        max_loan_amount = reg_approved.predict(user_df)[0]
        if requested_loan.value <= max_loan_amount:
            print(f"✅ Loan Approved! You are eligible for {requested_loan.value}. Maximum Loan Amount: {max_loan_amount:.2f}")
        else:
            print(f"✅ Loan Approved! But maximum you can get is {max_loan_amount:.2f}. Requested amount is too high.")
    else:
        max_loan_amount = reg_rejected.predict(user_df)[0]
        print(f"❌ Loan Rejected. You requested {requested_loan.value}, but the maximum you might get is {max_loan_amount:.2f}")

# Attach function to button
predict_button.on_click(predict_loan_status)

# Display widgets
display(gender, married, dependents, education, self_employed, applicant_income, 
        coapplicant_income, loan_term, credit_history, property_area, requested_loan, predict_button)
```
